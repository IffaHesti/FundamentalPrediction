```python
import os
import shutil

def get_directories(path):
    items = os.listdir(path)
    directories = [item for item in items if os.path.isdir(os.path.join(path, item))]
    return directories

current_directory = os.getcwd()

AllResults_path = f"{current_directory}/AllResults"
if not os.path.isdir(AllResults_path):
    os.makedirs(AllResults_path)

results_directories = get_directories(current_directory)

if "AllResults" in results_directories:
    results_directories.remove("AllResults")

for dir in results_directories:

    src = f"{current_directory}/{dir}/FinalResults"
    dst = f"{AllResults_path}/{dir}"

    if os.path.exists(src):  # cek dulu
        shutil.copytree(src, dst, dirs_exist_ok=True)
        print(f"Copied: {src}")
    else:
        print(f"Skipped (no FinalResults): {dir}")
```

```python
finalResults_path = f"{AllResults_path}/FinalResults"
if not os.path.isdir(finalResults_path):
    os.makedirs(finalResults_path)

# Get a list of all directory names
finalResults_directories = get_directories(AllResults_path)

if "FinalResults" in finalResults_directories:
    finalResults_directories.remove("FinalResults")

print(finalResults_directories)
```

```python
preprocessing_methods = ["MinMax", "Standard"]
target_columns = ["r_dicho", "b_dicho"]


bar_plot_path = f"{finalResults_path}/bar_plot"
if not os.path.isdir(bar_plot_path):
    os.makedirs(bar_plot_path)

KBest_rankedFeatures_path = f"{finalResults_path}/KBest_rankedFeatures"
if not os.path.isdir(KBest_rankedFeatures_path):
    os.makedirs(KBest_rankedFeatures_path)

for dir in finalResults_directories:
    tmp_filePath = f"{AllResults_path}/{dir}"
    tmp_fileName = f"bar_plot.png"
    shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{bar_plot_path}/{dir}.png")

    tmp_filePath = f"{AllResults_path}/{dir}"
    tmp_fileName = f"final_KBest_rankedFeatures.csv"
    shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{KBest_rankedFeatures_path}/{dir}.csv")

    for target in target_columns:

        max_path = f"{finalResults_path}/max/{target}"
        if not os.path.isdir(max_path):
            os.makedirs(max_path)

        mean_path = f"{finalResults_path}/mean/{target}"
        if not os.path.isdir(mean_path):
            os.makedirs(mean_path)

        auc_roc_mean_path = f"{finalResults_path}/auc_roc_mean/{target}"
        if not os.path.isdir(auc_roc_mean_path):
            os.makedirs(auc_roc_mean_path)

        aupr_mean_path = f"{finalResults_path}/aupr_mean/{target}"
        if not os.path.isdir(aupr_mean_path):
            os.makedirs(aupr_mean_path)

        for pp_method in preprocessing_methods:

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_max.csv"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{max_path}/{dir}.csv")

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_mean.csv"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{mean_path}/{dir}.csv")

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_{pp_method}_auc_roc_mean.png"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{auc_roc_mean_path}/{dir}_{pp_method}.png")

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_{pp_method}_aupr_mean.png"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{aupr_mean_path}/{dir}_{pp_method}.png")
```

```python
import pandas as pd

final_df = pd.DataFrame()

for dir in finalResults_directories:
    for target in target_columns:
        tmp_filePath = f"{AllResults_path}/{dir}"
        tmp_fileName = f"{target}_mean.csv"
        tmp_df = pd.read_csv(f"{tmp_filePath}/{tmp_fileName}")
        tmp_df.insert(0, "Target", f"{target}")
        tmp_df.insert(0, "Run_sample", f"{dir.split('_', 1)[1]}")
        final_df = pd.concat((final_df, tmp_df))

final_df = final_df.reset_index(drop=True)
final_df
# final_df.style.background_gradient(cmap="Reds", axis=False)
```

```python
# Function to extract mean and std components and convert to numeric types
def extract_and_convert(value):
    mean, std = map(float, value.split(' ± '))
    return mean, std

metrics = [
    'Acc Cv', 'Acc Te', 'Sp', 'Se',
    'MCC', 'F1', 'AUC-ROC', 'AUPR'
]

new_final_df = pd.DataFrame()
for i, col in enumerate(final_df.columns):
    if col not in metrics:
        new_final_df[col] = final_df[col].values.copy()

    else:
        new_final_df[[f"{col}_mean", f"{col}_std"]] = final_df[col].apply(extract_and_convert).apply(pd.Series)

new_final_df
```

```python
new_final_df = new_final_df.sort_values(by="AUPR_mean", ascending=False, ignore_index=True)
new_final_df.to_csv("final_df.csv", index=False)
new_final_df
```

```python
for target in target_columns:
    tmp_df = new_final_df.groupby("Target").get_group(target).reset_index(drop=True)
    tmp_df.drop(columns="Target", inplace=True)
    tmp_df.to_csv(f"{target}_df.csv", index=False)
```

```python
def table_generator(df_path, all=False):

    df = pd.read_csv(f"{df_path}_df.csv")

    # Rename and clean columns
    for col in df.columns:
        if "Acc Te" in col:
            df = df.rename(columns={col: col.replace("Acc Te_mean", "Acc")})
        if "_mean" in col:
            df = df.rename(columns={col: col.replace("_mean", "")})
        if "_std" in col:
            df = df.drop(columns=col)

    df = df[df["Alg"] != "DT"]  # Remove the 'DT' algorithm

    df = df[['Run_sample', 'Alg', 'Scaling', 'Acc', 'MCC', 'AUC-ROC', 'AUPR']]

    # Split 'Run_sample' into separate columns
    Run_sample = ['Fold', 'Sort', 'Group']
    df[Run_sample] = df['Run_sample'].str.split('_', expand=True)
    df = df.drop(columns=["Run_sample"])
    df['Sort'] = df['Sort'].str.replace('Sort', '')
    df['Group'] = df['Group'].str.replace('Group', '')

    df = df.sort_values(by=['AUPR', 'AUC-ROC'], ascending=[False, False], ignore_index = True)

    if all == True:
        df = df[['Alg', 'Fold', 'Group', 'Sort', 'Scaling', 'Acc', 'MCC', 'AUC-ROC', 'AUPR']]

        # Set the index
        df = df.set_index(['Alg', 'Fold', 'Group', 'Sort', 'Scaling'])

    else:

        # df = df[df['Sort'] != "company"]  # Filter out 'company' sort

        df = df[['Alg', 'Group', 'Acc', 'MCC', 'AUC-ROC', 'AUPR']]

        df = df.drop_duplicates(subset=['Alg','Group'], keep='first')

        # Calculate the maximum AUPR value for each algorithm group
        max_aupr_per_alg = df.groupby(by=['Alg'])['AUPR'].max().reset_index()
        max_aupr_per_alg = max_aupr_per_alg.rename(columns={'AUPR': 'Max_AUPR'})

        # Merge the max AUPR values back to the main DataFrame
        df = df.merge(max_aupr_per_alg, on=['Alg'])

        # Sort by 'Max_AUPR' and then by 'AUPR' within each algorithm group
        sorted_top_rows = df.sort_values(by=['Max_AUPR', 'AUPR'], ascending=[False, False])

        # Drop the Max_AUPR column as it was only used for sorting
        sorted_top_rows = sorted_top_rows.drop(columns='Max_AUPR', axis=1)

        # Prepare for further processing
        df = sorted_top_rows

        df['Group'] = df['Group'].replace('None', 'none')

        # Set the index
        df = df.set_index(['Alg', 'Group'])


    # # Save to CSV and LaTeX
    df.to_csv(f"{df_path}_table_top.csv" if all == False else f"{df_path}_table_all.csv")

    styled_df = df.style\
        .background_gradient(cmap="Oranges", subset=df.columns[0:]) \
        .format('{:.2f}', subset=df.columns[0:]) \
        .set_properties(**{'border': '1.3px solid black'})

    styled_df.to_latex(
        f"{df_path}_table_top.tex" if all == False else f"{df_path}_table_all.tex",
        clines="all;data",
        convert_css=True,
        position_float="centering",
        multicol_align="|c|",
        hrules=True,
    )
    display(styled_df)

table_generator("r_dicho")
table_generator("b_dicho")

# table_generator("r_dicho", all=True)
# table_generator("b_dicho", all=True)
```
